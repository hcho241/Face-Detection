```python
import os, sys
import cv2
import cvlib as cv
import numpy as np
```

```python
dirname = 'test'
# Create a list of images and a list of corresponding names 
(img, label, name, id) = ([], [], {}, 0) 
for (subdirs, dirs, files) in os.walk(dirname): 
    #print("subdirs :", subdirs)
    #print("dirs :", dirs)
    #print("files :", files)
    for subdir in dirs: 
        #print("subdir in dirs :", subdir)
        name[id] = subdir # name[0] = subdir 
        subjectpath = os.path.join(dirname, subdir) 
        for filename in os.listdir(subjectpath): 
            #print("filename", filename)
            path = subjectpath + '/' + filename 
            #print("path :", path)
            img.append(cv2.imread(path, 0)) 
            label.append(int(id)) 
        id += 1
#print("img", img)
#print("label", label)    
#print("name", name)
#print("id", id)

# Create a np array from the two lists above 
(img, label) = [np.array(lst) for lst in [img, label]] 
#print("np array :", (img, label))

# OpenCV trains a model from the images 
model = cv2.face.LBPHFaceRecognizer_create() 
model.train(img, label) 
#model.save("trainer.yml") # save model 

# Start webcam 
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 
webcam = cv2.VideoCapture(0) 
#(width, height) = (224, 224) 
(width, height) = (130, 100)
while True : 
    ret, frame = webcam.read() 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) 
    for (x, y, w, h) in faces: 
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2) 
        face = gray[y:y+h, x:x+w] 
        face_resize = cv2.resize(face, (width, height)) 
        # Try to recognize the face from webcam
        prediction = model.predict(face_resize) 
        print("predictions :", prediction)
        print("name :", name[prediction[0]])
        cv2.rectangle(frame, (x, y), (x+w, y+h),(255, 0, 0), 3) 
        if 80 <= prediction[1] : 
            answer = "{} {:.2f}%".format(name[prediction[0]], prediction[1])
            cv2.putText(frame, answer, (x-10, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
        else : 
            cv2.putText(frame, 'not recognized', (x-10, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0)) 
    # Display the resulting frame
    cv2.imshow('Webcam', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): # quit 
        break
webcam.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import os
```

```python
webcam = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

count = 0
face_data = []
dataset_path = './knn_openCV/'
offset = 10

file_name = input("Enter the name of the person :  ")
while True:
    ret,frame = webcam.read()
    
    if(ret == False):
        continue
        
    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)    
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    #print('before sorting', faces)
    faces = sorted(faces, key=lambda f:f[2]*f[3])
    #print('after sorting', faces)
    
    # pick the last face (largest)
    for face in faces[-1:]:
        x, y, w, h =  face 
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255,0,0), 2)
    
    # extract only face in different window 
    face_section = frame[y-offset:y+h+offset, x-offset:x+w+offset]
    face_section = cv2.resize(face_section,(100,100))
    print('face section', face_section)
    count += 1
    print('count', count)
    if(count % 10 == 0):
        face_data.append(face_section)
        print('face data length', len(face_data))      
    cv2.imshow("FACE CROP",face_section) 
    cv2.imshow("WEBCAM",frame)
        
    if cv2.waitKey(1) & 0xFF == ord('q'):
            break
webcam.release()
cv2.destroyAllWindows()        
# convert our face list array into a numpy array
face_data = np.array(face_data)
#print("face shape ", face_data.shape)
#print("face shape[0] ", face_data.shape[0])
face_data = face_data.reshape((face_data.shape[0],-1)) 
#print("reshape ", face_data.shape)

#save this data into numpy array file & text file 
np.save(dataset_path + file_name + '.npy', face_data)
print("data successfully saved at " + dataset_path + file_name + '.npy')
```

```python
########## KNN CODE ############
def distance(v1, v2):
    # Eucledian 
    return np.sqrt(((v1-v2)**2).sum())

def knn(train, test, k=5):
    dist = []
    for i in range(train.shape[0]):
        # Get the vector and label
        ix = train[i, :-1]
        iy = train[i, -1]
        # Compute the distance from test point
        d = distance(test, ix)
        #print('eucledian dist ', d)
        dist.append([d, iy])
        #print('dist list ', dist)
    # Sort based on distance and get top k
    dk = sorted(dist, key=lambda x: x[0])[:k]
    #print('dk', dk)
    # Retrieve only the labels
    labels = np.array(dk)[:, -1]
    #print('labels', labels)
    # Get frequencies of each label
    output = np.unique(labels, return_counts=True)
    #print('output', output)
    # Find max frequency and corresponding label
    index = np.argmax(output[1])
    return output[0][index]
################################

#Init webcam
webcam = cv2.VideoCapture(0)
# Face Detection
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
skip = 0
dataset_path = './knn_openCV/'
face_data = [] 
labels = []
class_id = 0 # Labels for the given file
names = {} # Mapping btw id - name

# Data Preparation
for dataset in os.listdir(dataset_path):
    #print("Loaded "+ dataset)
    if dataset.endswith('.npy'):
        # Create a mapping btw class_id and name
        names[class_id] = dataset[:-4] # extract .npy part in filename
        data_item = np.load(dataset_path + dataset)
        face_data.append(data_item)
        # Create ;abels for the class
        target = class_id * np.ones((data_item.shape[0],))
        class_id += 1
        labels.append(target)
#print('face data', face_data)
#print('face labels', labels)
# join a sequence of arrays 
face_dataset = np.concatenate(face_data, axis=0)
face_labels = np.concatenate(labels, axis=0).reshape((-1, 1))
trainset = np.concatenate((face_dataset, face_labels), axis=1)
#print('trainset shape ', trainset.shape)

# Testing 
while True:
    ret,frame = webcam.read()
    if (ret == False):
        continue
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    if(len(faces) == 0):
        continue
    for face in faces:
        x, y, w, h = face
        # Get the face ROI
        offset = 10
        face_section = frame[y-offset:y+h+offset, x-offset:x+w+offset]
        face_section = cv2.resize(face_section,(100,100))
        # Predicted Label 
        prediction = knn(trainset, face_section.flatten())
        #print('prediction knn ', prediction)
        # Display on the screen the name and bounding box 
        #print('names ', names)
        pred_name = names[int(prediction)]
        cv2.putText(frame, pred_name, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
    cv2.imshow("Faces",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
            break
webcam.release()
cv2.destroyAllWindows()  
```
